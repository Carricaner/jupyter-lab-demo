```python
import matplotlib.pyplot as plt
```

```python
# Step 2: Prepare the data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]   # Line 1 data
y2 = [1, 3, 6, 10, 15]   # Line 2 data
y3 = [1, 2, 3, 4, 5]     # Line 3 data

# Step 3: Plot the lines
plt.plot(x, y1, label='Line 1', color='r')  # Red line
plt.plot(x, y2, label='Line 2', color='g')  # Green line
plt.plot(x, y3, label='Line 3', color='b')  # Blue line

# Step 4: Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Three Lines Plot')
plt.legend()

# Step 5: Display the plot
plt.show()
```
