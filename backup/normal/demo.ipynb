```python
import matplotlib.pyplot as plt
```

```python
def calculate(total_money: int, rate: float, years: int):
    year_money = round(total_money / years)
    snapshot = 0
    result = []
    for year in range(years):
        snapshot = round((snapshot + year_money) * (1 + rate))
        result.append(snapshot - year_money * (1 + year))
    return result
```

```python
calculate(1_000_000, 0.05, 10)
```

```python
years = 10
budget = 2_000_000
yield_rate = 0.06

# x = [year + 1 for year in range(years)]
# y1 = calculate(budget, yield_rate, years)

# plt.plot(x, y1, label='Line 1')

years = 5
x = [year + 1 for year in range(years)]
y2 = calculate(budget, yield_rate, years)

plt.plot(x, y2, label='Line 2')

years = 2
x = [year + 1 for year in range(years)]
y3 = calculate(budget, yield_rate, years)

plt.plot(x, y3, label='Line 3')

for i in range(len(x)):
    if x[i] % 1 == 0:
        # plt.annotate(f'{y1[i]}', (x[i], y1[i]), textcoords="offset points", xytext=(0,10), ha='center')
        plt.annotate(f'{y2[i]}', (x[i], y2[i]), textcoords="offset points", xytext=(0,10), ha='center')
        plt.annotate(f'{y3[i]}', (x[i], y3[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Step 4: Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Three Lines Plot')
plt.legend()

# Add grid
plt.grid(True)

# Step 5: Display the plot
plt.show()
```
